 # Converting data types

```python
tips = pd.read_csv('tips.csv')
print (tips.head())
```

```python
tips.sex = tips.sex.astype('category')
```

```python
tips['total_bill'] = pd.to_numeric(tips['total_bill'], errors='coerce')
```

 # Custom functions to clean data

```python
import numpy as np

def recode_sex(sex_values):
    if sex_values == 'Male':
        return 1
    elif sex_values == 'Female':
        return 0
    else:
        return np.nan
    
tips['recode_sex']=tips.sex.apply(recode_sex)
```

```python
print (tips.head())
```

# Lambda functions

```python
tips['total_dollar_replace'] = tips['total_dollar'].apply(lambda x:x.replace('$',''))
```
